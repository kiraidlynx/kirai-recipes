# Clustering Completo

### Análisis exploratorio

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
```

```python
#Generación de datos aleatorios
np.random.seed(4711)
a = np.random.multivariate_normal([10,0],[[3,1],[1,4]], size = [100,])
b = np.random.multivariate_normal([0,20], [[3,1],[1,4]], size = [50,])
X = np.concatenate((a,b))

print(X.shape)
plt.scatter(X[:,0], X[:,1])
```

```python
Z=linkage(X,"ward")
```

```python
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist
```

```python
c, coph_dist = cophenet(Z, pdist(X))
# c porcentaje de conservación de las distancias originales (~1 clustering perfecto)
```

```python
## Exploración de puntos iniciales
idx = [33,62,68]
idx2 = [15,69,41]
plt.figure(figsize=(10,8))
plt.scatter(X[:,0], X[:,1])##pintar todos los puntos
plt.scatter(X[idx,0], X[idx,1], c='r')##destacamos en rojo los puntos interesantes
plt.scatter(X[idx2,0], X[idx2,1], c='y')##destacamos en amarillo el segundo cluster
plt.show()
```

### Representación gráfica

```python
plt.figure(figsize=(25,10))
plt.title("Dendrograma del clustering jerárquico")
plt.xlabel("Índices de la Muestra")
plt.ylabel("Distancias")
dendrogram(Z, leaf_rotation=90., leaf_font_size=8.0, color_threshold=0.7*180) #labels=[array_etiquetas]
plt.show()
```

### Truncar el dendrograma

```python
plt.figure(figsize=(25,10))
plt.title("Dendrograma del clustering jerárquico truncado")
plt.xlabel("Índices de la Muestra")
plt.ylabel("Distancias")
dendrogram(Z, leaf_rotation=90., leaf_font_size=12.0, color_threshold=0.7*180,
           truncate_mode="lastp", p=12, show_leaf_counts=True, show_contracted=True,)
#lastp=últimos elementos p unidos, p=elementos

plt.show()
```

#### ~Función personalizada dendograma

```python
def dendrogram_tune(*args, **kwargs):
    
    max_d=kwargs.pop("max_d", None)
    if max_d and 'color_threshold' not in kwargs:
        kwargs['color_threshold'] = max_d
    annotate_above = kwargs.pop('annotate_above', 0)
    
    ddata = dendrogram(*args,**kwargs)
        
    if not kwargs.get('no_plot', False):
        plt.title("Clustering jerárquico con Dendrograma truncado")
        plt.xlabel("Índice del Dataset (o tamaño del cluster)")
        plt.ylabel("Distancia")
        for i, d, c in zip(ddata['icoord'], ddata['dcoord'], ddata['color_list']):
            x = 0.5 * sum(i[1:3])
            y = d[1]
            if y>annotate_above:
                plt.plot(x,y,'o',c=c)
                plt.annotate('%.3g'%y, (x,y), xytext=(0,-5),
                            textcoords="offset points", va="top", ha="center")
                
    if max_d:
        plt.axhline(y=max_d, c='k')
        
    return ddata
```

```python
dendrogram_tune(Z,truncate_mode='lastp',p=12, leaf_rotation=90., leaf_font_size=12.,
                show_contracted=True,annotate_above=10, max_d=50)
plt.show()
```

### Corte automático del dendrograma

#### ~Método del codo

```python
last = Z[-10:,2] #distancias de los últimos elementos (alturas)
last_rev = last[::-1] ##girarlos (el último es el primero)
print(last_rev)
idx = np.arange(1, len(last)+1)
plt.plot(idx, last_rev)

acc = np.diff(last,2) ##aceleración: diferencia de los cambios de 2 en dos
acc_rev = acc[::-1] ##girarlos
plt.plot(idx[:-2]+1, acc_rev) #todos menos los dos últimos y lo desplazamos 1
plt.show()
k = acc_rev.argmax() +2 ## máximo de la aceleración
print("El número óptimo de cluster es %s"%str(k))
```

No decidir únicamente en base a la regla del codo. Hay que condicionar con conocimiento de la causa del cluster.

### Recuperar clusters y elementos

```python
from scipy.cluster.hierarchy import fcluster
```

```python
max_d=25 #corte por distancia de corte (altura)
clusters = fcluster(Z, max_d, criterion="distance")
clusters
```

```python
k=2 #corte por número de clusters
clusters = fcluster(Z, k, criterion="maxclust")
clusters
```
